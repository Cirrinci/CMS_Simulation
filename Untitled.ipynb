```python
def calculate_mean_std(column):
    mean = column.mean()
    std = column.std()
    return pd.Series({'mean': mean, 'std': std})

grouped_stats = df.groupby(['runId', 'eventId'])['genpart_et'].apply(calculate_mean_std)

# Display the result
print(grouped_stats)
```

```python
def calculate_deltaR(group):
    # Calculate deltaR for pairs (1, 2), (3, 4), (5, 6), and (7, 8) within each group
    deltas = []
    for i in range(0, 7, 2):  # Iterate over indices 0, 2, 4, and 6
        eta_diff = np.sqrt((group.iloc[i]['seed_eta'] - group.iloc[i+1]['seed_eta'])**2
                           +(group.iloc[i]['seed_phi'] - group.iloc[i+1]['seed_phi'])**2)
        deltas.append(eta_diff)
    return pd.Series(deltas, index=['deltaR_1_2', 'deltaR_3_4', 'deltaR_5_6', 'deltaR_7_8'])

grouped_deltas = df.groupby(['runId', 'eventId']).apply(calculate_deltaR)

```
